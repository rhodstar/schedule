```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
def get_link_carreras(plan):
    lista_de_carreras = []
    carreras = soup.find_all('a',class_='boton_animado',text=plan)
    for i in carreras:
        lista_de_carreras.append(url_carreras+i['href'])
    return lista_de_carreras

def get_link_siae(url):
    html_siae = requests.get(url).text
    sopa = BeautifulSoup(html_siae)
    lista_siae = []
    
    siae = sopa.find_all('a',{'target':'_blank'})
    for i in siae:
        if 'dgae-siae' in i['href']:
            lista_siae.append(i['href'])
            break
    return lista_siae
#-----------------------------------------
url_carreras = 'http://escolar.ingenieria.unam.mx/mapas/'
html_carreras = requests.get(url_carreras).text
soup = BeautifulSoup(html_carreras)

lista_de_carreras = []    
for i in ['Plan 2016']:#['Plan 2010','Plan 2016','Plan 2020']:
    lista_de_carreras.extend(get_link_carreras(i))
links_siae = []
for i in lista_de_carreras:
    links_siae.extend(get_link_siae(i))
with open('links_siae.txt', 'w') as f:
    f.write("\n".join(links_siae))
```

```python
def claves_de_asignaturas(url):
    html = requests.get(url).text
    soup = BeautifulSoup(html)
    
    try:
        titulo = soup.find('td',class_="CellTit").text
        #titulo += soup.find_next('td',class_="CellTit").text
    except:
        titulo = 'null'
    claves = soup.find_all('td',class_="CellDat",attrs={'align':'left'})
    lista_de_llaves = []
    lista_de_valores = []
    for value in claves:
        if ('OPTATIVA' not in value.text ):
            if ('OBLIGATORIA' not in value.text):
                lista_de_llaves.append(value.text)
    claves = soup.find_all('input',{"name":"asg"})
    for i in claves:
        lista_de_valores.append(i['value'])
    diccionario = dict(zip(lista_de_valores,lista_de_llaves))
    return titulo,diccionario

diccionario_por_carrera = {}
for i in links_siae:
    llave,valor = claves_de_asignaturas(i)
    diccionario_por_carrera[llave] = valor
```

```python
def Merge(dict1, dict2): 
    return(dict2.update(dict1)) 
diccionario_claves_nombres = {}
for key in diccionario_por_carrera:
    #print(diccionario_por_carrera[key])
    Merge(diccionario_por_carrera[key],diccionario_claves_nombres)
    
dataAlgo = pd.Series(diccionario_claves_nombres)
dataAlgo = pd.DataFrame(dataAlgo)
dataAlgo.to_csv("claves_materias_nombres.csv")
```

```python
subject = """Subject::create([
    'id' => {},
    'name' => '{}'
]);\n
"""
Subject_Seeder = ''
for key in diccionario_claves_nombres:
    Subject_Seeder += subject.format(key,diccionario_claves_nombres[key])

with open('Subject_Seeder.php', 'w') as f:
    f.write(Subject_Seeder)
```
